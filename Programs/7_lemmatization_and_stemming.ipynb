## Stemming

```python
with open("Resources/StemLem.txt") as f:
    text=f.readlines()
text
```

```python
text=" ".join(text)
text
```

```python
from nltk.tokenize import word_tokenize
word=word_tokenize(text)
word
```

## Porter stemmer

```python
from nltk.stem.porter import PorterStemmer
port=PorterStemmer()

for i in word:
    print(i, "--> "+ port.stem(i))
```

## Snowball stemmer

```python
from nltk.stem.snowball import SnowballStemmer

snow=SnowballStemmer('english')

```

```python
for i in word:
    print(i, "--> "+snow.stem(i))
```

## Lancaster stemmer

```python
from nltk.stem.lancaster import LancasterStemmer

lan=LancasterStemmer()
```

```python
for i in word:
    print(i," --> "+lan.stem(i))
```

## Lemmatization

## Spacy lemmatization

```python
import spacy
nlp=spacy.load('en_core_web_sm')

doc=nlp(text)

for i in doc:
    print(i,"---> "+i.lemma_)
```

## Wordnet lemmatizer

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
wl=WordNetLemmatizer()
for i in text.split():
    print(i,"---> "+wl.lemmatize(i))
```

## Text blob

```python
from textblob import TextBlob
```

```python
tb=TextBlob(text)
```

```python
for i in tb.words:
    print(i, "---> "+i.lemmatize())
```
